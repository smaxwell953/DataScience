The column is a mix of integers and floats, so let's change the type of column to float.

```python
menu_no_nan['Total Fat (g)'] = menu_no_nan['Total Fat (g)'].astype(float)
```

```python
menu_no_nan.info()
```

Now the dtype of the total fat column is float. Let's look at the other columns with dtypes of 'object'.

The columns 'Vitamin A (% DV)', 'Vitamin C (% DV)', 'Calcium (% DV)', 'Iron (% DV)' have percentages in their columns. Let's convert these values to floats for consistency.

```python
menu_no_nan['Vitamin A (% DV)'] = menu_no_nan['Vitamin A (% DV)'].str.rstrip('%').astype('float')
```

```python
menu_no_nan['Vitamin C (% DV)'] = menu_no_nan['Vitamin C (% DV)'].str.rstrip('%').astype('float')
```

```python
menu_no_nan['Calcium (% DV)'] = menu_no_nan['Calcium (% DV)'].str.rstrip('%').astype('float')
```

```python
menu_no_nan['Iron (% DV)'] = menu_no_nan['Iron (% DV)'].str.rstrip('%').astype('float')
```

The caffeine column also has mixed types, integers and strings ('Varies' and 'varies'). Let's check out the rows with string values.

```python
menu_varies = menu_no_nan[(menu_no_nan['Caffeine (mg)'] == 'Varies') | (menu_no_nan['Caffeine (mg)'] == 'varies')]
print(len(menu_varies))
print(round((len(menu_varies)/len(menu)),3))
```

```python
menu_varies
```

<p>There are 22 rows with 'varies' or 'Varies' in the caffeine column. With such a small dataset, dropping 9% of the items wouldn't be a good idea, so let's find a way to work with these 22 rows.</p>

<p>To make the values consistent, we would need to convert the 'varies' and 'Varies' to an integer. Let's see the minimum and maximum values of the items with integers in the Caffeine column.</p>

```python
menu_no_varies = menu_no_nan[(menu_no_nan['Caffeine (mg)'] != 'Varies') & (menu_no_nan['Caffeine (mg)'] != 'varies')]
menu_no_varies
```

```python
menu_no_varies_min = menu_no_varies['Caffeine (mg)'].min()
menu_no_varies_min
```

```python
menu_no_varies_max = menu_no_varies['Caffeine (mg)'].max()
menu_no_varies_max
```

The minimum value is 0 and the maximum value is 95, so let's use 500 to represent the 'varies' and 'Varies' menu items.

```python
menu_no_nan['Caffeine (mg)'] = menu_no_nan['Caffeine (mg)'].replace(['varies','Varies'],500)
```
