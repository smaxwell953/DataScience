```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
data = pd.read_csv('Counties2016.csv')
```

```python
data.head(5)
```

```python
data['2012Total'] = data['2012D'] + data['2012R'] + data['2012O']
data['2012DPCT'] = data['2012D'] / data['2012Total']
data['2012RPCT'] = data['2012R'] / data['2012Total']
data['2012OPCT'] = data['2012O'] / data['2012Total']
```

```python
data['2012Margin'] = data['2012RPCT'] - data['2012DPCT']
```

```python
data['2016Total'] = data['2016D'] + data['2016R'] + data['2016O']
data['2016DPCT'] = data['2016D'] / data['2016Total']
data['2016RPCT'] = data['2016R'] / data['2016Total']
data['2016OPCT'] = data['2016O'] / data['2016Total']
```

```python
data['2016Margin'] = data['2016RPCT'] - data['2016DPCT']
data['Diff_12_16'] = data['2016Margin'] - data['2012Margin']
```

```python
#export cleaned data as CSV
data.to_csv('2020_cleaned.csv')
```
