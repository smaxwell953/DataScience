```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
votes = pd.read_csv('./ElectionData/Michigan.csv', low_memory=False)
votes.head()
```

```python
votes.info()
```

```python
votes = votes.fillna(0)
```

```python
votes['Year'] = pd.to_numeric(votes['Year'])
```

```python
votes.loc[:,'Total Votes'] = (votes.sum(axis=1)) - votes['Year']
```

```python
pres = votes[votes['Office'] == 'President']
```

```python
pres = pres.drop('Office', axis = 1)
```

```python
pres['Year'] = pd.to_numeric(pres['Year'])
```

```python
pres['PCT_DEM'] = (round(pres['D#']/pres['Total Votes'],4))*100
```

```python
pres['PCT_REP'] = (round(pres['R#']/pres['Total Votes'],4))*100
```

```python
pres['RD_DIFF'] = pres['PCT_REP'] - pres['PCT_DEM']
```

```python
pres_pct = pres[['State', 'Year', 'County', 'RD_DIFF']]
```

```python
pres_pct_2000 = pres_pct[pres_pct['Year'] >= 2000]
```

```python
counties = ['Alger', 'Baraga', 'Chippewa', 'Delta', 'Dickinson', 'Gogebic', 'Houghton', 'Iron', 'Keweenaw', 'Luce', \
            'Mackinac', 'Marquette', 'Menominee', 'Ontonagon', 'Schoolcraft']
```

```python
for c in counties:
    print(pres_pct_2000[pres_pct_2000['County'] == c])
```
