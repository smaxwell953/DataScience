```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
votes = pd.read_csv('./ElectionData/California.csv', low_memory=False)
votes.head()
```

```python
votes.info()
```

```python
votes.loc[:, votes.dtypes == np.object]
```

```python
votes = votes.drop(columns = ['State', 'Candidate 1', 'Candidate 2', 'Candidate 3', 'Candidate 4', 'Candidate 5'])
```

```python
votes.loc[:,'Total Votes'] = votes.sum(axis=1)
```

```python
pres = votes[votes['Office'] == 'President']
```

```python
pres = pres.drop('Office', axis = 1)
```

```python
pres['Year'] = pd.to_numeric(pres['Year'])
```

```python
years = pres['Year'].unique()
df1 = pres[['Year', 'County', 'Total Votes']]

pct_df = pd.DataFrame([])

for y in years:
    df = df1[df1['Year'] == y]
    df['State Total'] = df['Total Votes'].sum()
    df['County %'] = (round((df['Total Votes']) / (df['State Total'])*100, 2))
    pct_df = pct_df.append(df.sort_values(by='County %', ascending=False))
```

```python
pct_df = pct_df.set_index('Year')
```

```python
pct_df = pct_df[['County', 'County %']]
```

```python
pct_df.to_csv('pct_df.csv')
```
